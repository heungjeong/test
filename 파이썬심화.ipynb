```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
from urllib.parse import quote_plus
 
baseUrl = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
plusUrl = input('검색어 입력: ') 
crawl_num = int(input('크롤링할 갯수 입력(최대 50개): '))
 
url = baseUrl + quote_plus(plusUrl) 
html = urlopen(url)
soup = bs(html, "html.parser")
img = soup.find_all('img')
#print(soup.find_all('img', limit=2))
n = 1
for i in img:
    print(n)
    imgUrl = i.attrs['src']
    with urlopen(imgUrl) as f:
        with open('img' + str(n)+'.jpg','wb') as h: # w - write b - binary
            h.write(f.read())
    n += 1
    if n > crawl_num:
        break 
    
print('Image Crawling is done.')
```

```python
#유튜브 검색 목록
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome('chromedriver')
url = "https://www.youtube.com"
driver.implicitly_wait(3)
driver.get(url)
time.sleep(1)   #웹페이지가 열리는 시간을 고려하여 1초의 delay를 줍니다.
keyword = "피아노곡"
elem = driver.find_element_by_name("search_query")
elem.send_keys(keyword)
elem.send_keys(Keys.RETURN)
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
tags = soup.find_all('ytd-video-renderer', class_='style-scope ytd-item-section-renderer')

rank = 1
for i in tags:
    title = i.find(class_='yt-simple-endpoint style-scope ytd-video-renderer')['title']
    name = i.find('a', class_='yt-simple-endpoint style-scope yt-formatted-string').text
    name = name.replace('\n', '')
    link = i.find('a', class_='style-scope ytd-video-renderer')['href']
    link = 'https://www.youtube.com'+link
    print('순위 : ' + str(rank) + "\n제목 :" + title + "\n채널 : " + name + "\n링크 : " + link + "\n")
    #print(f'순위 : {rank}\n제목 : {title}\n채널 : {name}\n링크 : {link}\n')
    rank += 1
driver.close()
```

```python
def divide(x,y):
    try:
        z=x/y
    except ZeroDivisionError:
        print("0으로 나누면 에러가 발생합니다.")
    else:
        print(z)
    finally:
        print("divide 함수 종료")
        
divide(10,2)
divide(10,0)
a,b = map(int, input().split())
divide(a,b)
    
```
